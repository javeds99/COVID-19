```python
sub.head()
```

```python
#3. Create a correlation matrix heatmap (colored).
sub.corr()
```

```python
f,ax = plt.subplots(figsize = (10,10))
```

```python
corr = sns.heatmap(sub.corr(), annot = True, ax = ax) 
```

```python
corr.figure
```

Anything from 0.58-0.70 represents a strong negative correlation, and are shown as the darker squares.
Anything from 0.71-0.80 represents a weak positive correlation, and are shown as the dark red-orange squares.
Anything from 0.81-1 represents a strong positive correlation, and are shown as the bright orange-beige squares.


```python
#4. Create a scatterplot for the variables with the strongest correlations and examine the nature of their relationships.
sns.lmplot (x= 'Cumulative Cases', y= 'Cumulative Deaths', data=sub)
```

This scatterplot shows a mid-level to high positive relationship.
